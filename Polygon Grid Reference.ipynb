```python
# Dependencies
import numpy as np
import requests
import json
import pandas as pd
from datetime import datetime as dt
```

```python
# Incorporate list of points
grids = "GRIDS.xlsx"
grids_df = pd.read_excel(grids)
grids_df.head()
```

```python
#Create URL list

urls = []
posts = []

for g in range(0,len(grids_df),1):
    
    try:
        
        office = "OKX"
        
        gridX = grids_df['gridX'][g]
        gridY = grids_df['gridY'][g]
        
        urls.append(f'https://api.weather.gov/gridpoints/{office}/{gridX},{gridY}')
        posts.append(requests.get(urls[g]).json())
    
    except:
        pass
    
    print(f'{urls[g]} : succesfully documented')
    print("-----------------------------")
    
print("URLs Logged")
print("-----------------------------")
```

```python
# URL for GET requests to retrieve data

base_urls = []

for g in range(0,len(urls),1):

    try:

        base_urls.append(urls[g])

    except:
        pass

#Set responses

responses = []

for b in range(0,len(base_urls),1):

    try:
            
        responses.append(requests.get(base_urls[b]).json())

    except:
        pass

#Align dictionaries

weather = []
grids = []

for r in range(0,len(responses),1):

    try:

        weather.append(responses[r]['geometry'])
        grids.append(responses[r]['properties'])

    except:
        pass        
```

```python
# Assign coordinates to each polygon 

polys = []
p_list = []

plot_x = []
plot_y = []

dfs = []

for z in range(0,len(weather),1):
    
    polys.append(weather[z]['coordinates'])
    
    #print(f"{z+1}")
    #print("-----------------------")
    #print(f"{polys[z]}")
    #print("-----------------------")

for l in range(0,len(polys),1):
    
    p_list.append(polys[l][0])

    #print(f"{l+1}")
    #print("-----------------------")
    #print(f"{p_list[l]}")
    #print("-----------------------")
    
for h in range(0,len(weather),1):
    
    for s in range(0,5,1):
        
        plot_x.append(p_list[h][s][0])
        plot_y.append(p_list[h][s][1])
        
        #print(f"{grids[h]['gridX']},{grids[h]['gridY']}")
        #print(f"{p_list[h][s][0]} | {p_list[h][s][1]}")
        #print("----------------------------")
        
        row = {
            'gridX': grids[h]['gridX'],
            'gridY': grids[h]['gridY'],
            'x_coordinate': p_list[h][s][0],
            'y_coordinate': p_list[h][s][1]
        }
        # Append the dictionary to the data list
        dfs.append(row)

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(dfs)

# Save the DataFrame to a CSV file
csv_file_path = 'grids.csv'  # Specify your desired file path and name
df.to_csv(csv_file_path, index=False)
```
